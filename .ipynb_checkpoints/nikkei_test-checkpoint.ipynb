```python
import torch.nn as nn
import torch
from torch.optim import SGD
from torch.optim import Adam
import math
import numpy as np
import csv
import matplotlib.pyplot as plt
import torch.utils.tensorboard as tensorboard
from torch.autograd import Variable  
import datetime
```

```python
#1時間ごとでの
with open('2022_nikkei_225.csv') as f:
    reader = csv.reader(f)
    l = [row for row in reader]
    hours_data = [int(l[i*60][2]) for i in range(2, len(l) // 60)]
    h_data = np.array(hours_data)
```

```python
def normalization(data):
    avg = np.mean(data)
    std = np.std(data)
    data = (data - avg) / std
    return data
```

```python
h_data = normalization(h_data)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot(h_data, color = "r")
```

```python
window = 10 # 移動平均の範囲
w = np.ones(window)/window

h_data_convolve = np.convolve(h_data, w, mode='same')

fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot(h_data_convolve, color = "r")
```

```python
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
print(torch.cuda.is_available())
```

```python
##隣のデータとの差の平均
sum = 0
for i in range(len(h_data)-1):
#     sum += np.sqrt((h_data[i+1]-h_data[i])**2)
    sum += np.sqrt((h_data_convolve[i+1]-h_data_convolve[i])**2)
print(sum/len(h_data))
```

```python
def mkDataSet(data_size, data_length):
    train_x = []
    train_t = []

    for offset in range(data_size):
#         train_x.append([[h_data[offset+i]] for i in range(data_length)])
#         train_t.append([h_data[offset+data_length]])
        train_x.append([[h_data_convolve[offset+i]] for i in range(data_length)])
        train_t.append([h_data_convolve[offset+data_length]])

    return train_x, train_t #train_x=(data_size, data_length, 1), train_t=(data_size, 1)
```

```python
class Predictor(nn.Module):
    def __init__(self, inputDim, hiddenDim, outputDim):
        super(Predictor, self).__init__()
        self.hiddenDim = hiddenDim
        self.kernel_size1 = 10
        self.kernel_size2 = 5
        
        self.rnn = nn.LSTM(input_size=inputDim, hidden_size=hiddenDim, batch_first=True) #batch_first=Trueで(seq, batch, vec)->(batch, seq, vec)に入力の形を変更
        self.conv1d_1 = nn.Conv1d(1, 3, self.kernel_size1)
        self.conv1d_2 = nn.Conv1d(3, 1, self.kernel_size2)
        self.output_layer = nn.Linear((hiddenDim - self.kernel_size1 + 1) - self.kernel_size2 + 1, outputDim)

    def forward(self, inputs):
        h_0 = Variable(torch.zeros(1, 1, self.hiddenDim).to(device))
        c_0 = Variable(torch.zeros(1, 1, self.hiddenDim).to(device))
        output, (hidden, cell) = self.rnn(inputs, (h_0, c_0)) #LSTMのforwardのreturnはこのような戻り値になっている
        output = self.conv1d_1(output[:, -1, :])
        output = self.conv1d_2(output)
        output = self.output_layer(output) #LSTMのoutput=(batch, seq, hidden)からseqのみ最後のやつだけを取り出す
        return output
```

```python
data_length = 50
training_size = len(h_data)-data_length
epoch_num = 100
hidden_size = 20

train_x, train_t = mkDataSet(training_size, data_length)

model = Predictor(1, hidden_size, 1).to(device)
criterion = nn.MSELoss()
optimizer = SGD(model.parameters(), lr=0.01)
```

```python
last_training_accuracy = 0.0
for epoch in range(epoch_num):
    running_loss = 0.0
    correct = 0.0
    for i in range(training_size):
        optimizer.zero_grad()
        data = torch.tensor([train_x[i]]).float().to(device)
        label = torch.tensor([train_t[i]]).float().to(device)

        output = model(data)

        loss = criterion(output, label)
        loss.backward()
        optimizer.step()

        running_loss += loss.data
        correct += np.sum(np.abs((output.data - label.data).cpu().numpy()) < 0.03)
    
    training_accuracy = correct / training_size
    print('%d loss: %3f, training_accuracy: %.5f' % (epoch+1, running_loss, training_accuracy))
    if last_training_accuracy > training_accuracy:
        break
    last_training_accuracy = training_accuracy
```
