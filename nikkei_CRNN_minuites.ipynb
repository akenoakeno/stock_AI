```python
import tqdm
import torch.nn as nn
import torch
from torch.optim import SGD
from torch.optim import RMSprop
from torch.optim import Adam
import math
import numpy as np
import csv
import matplotlib.pyplot as plt
import torch.utils.tensorboard as tensorboard
from torch.autograd import Variable  
import datetime
```

```python
#1時間ごとでの
with open('2022_nikkei_225.csv') as f:
    reader = csv.reader(f)
    l = [row for row in reader]
    minuites_data = [int(l[i][2]) for i in range(2, len(l))]
    m_data = np.array(minuites_data)
```

```python
def normalization(data):
    avg = np.mean(data)
    std = np.std(data)
    data = (data - avg) / std
    return data
```

```python
m_data = normalization(m_data)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot(m_data, color = "r")
```

```python
window = 10 # 移動平均の範囲
w = np.ones(window)/window

m_data_convolve = np.convolve(m_data, w, mode='same')

fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot(m_data_convolve, color = "r")
```

```python
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
print(torch.cuda.is_available())
```

```python
def value_s(np_array, rate = 0.9):
    sum = 0
    if (len(np_array) < 100):
        for n, r in enumerate(np_array):
            sum += rate ** n * r
    else:
        for n in range(100):
            sum += rate ** n * np_array[n]
    return sum

def mkDataSet(data_size, data_length, batch_size):
    train_x = []
    train_t = []

    for offset in range(data_size):
        train_x.append([[m_data_convolve[offset+i]] for i in range(data_length)])
        train_t.append([value_s(m_data_convolve[offset+data_length:-1])])
        
    training_size = data_size // batch_size - 1
    train_x = np.array(train_x[0:training_size]).reshape(-1, batch_size, data_length)
    train_t = np.array(train_t[0:training_size]).reshape(-1, batch_size)

    return train_x, train_t, training_size
```

```python
class Predictor(nn.Module):
    def __init__(self, inputDim, hiddenDim, outputDim):
        super(Predictor, self).__init__()
        self.hiddenDim = hiddenDim
        self.kernel_size1 = 200
        self.kernel_size2 = 50
        lstm_input_size = (inputDim - self.kernel_size1 + 1) - self.kernel_size2 + 1
        
        self.conv1d_1 = nn.Conv1d(1, 3, self.kernel_size1)
        self.conv1d_2 = nn.Conv1d(3, 1, self.kernel_size2)
        self.rnn = nn.LSTM(input_size = lstm_input_size, hidden_size=hiddenDim, batch_first=True, bidirectional=True) #batch_first=Trueで(seq, batch, vec)->(batch, seq, vec)に入力の形を変更
        self.output_layer = nn.Linear(hiddenDim, outputDim)

    def forward(self, inputs):
        h_0 = Variable(torch.zeros(1, batch_size, self.hiddenDim).to(device))
        c_0 = Variable(torch.zeros(1, batch_size, self.hiddenDim).to(device))        
        outputs = self.conv1d_1(inputs)
        outputs = self.conv1d_2(outputs)
        output, (hidden, cell) = self.rnn(inputs, (h_0, c_0)) #LSTMのforwardのreturnはこのような戻り値になっている

        output = self.output_layer(output) #LSTMのoutput=(batch, seq, hidden)からseqのみ最後のやつだけを取り出す
        return output
```

```python
data_length = 1000
data_size = len(m_data)-data_length
epoch_num = 100
hidden_size = 20
batch_size = 32

train_x, train_t, training_size = mkDataSet(data_size, data_length, batch_size)
```

```python
model = Predictor(1, hidden_size, 1).to(device)
criterion = nn.MSELoss()
optimizer = RMSprop(model.parameters(), lr=0.0005)
```

```python
for epoch in range(epoch_num):
    running_loss = 0.0
    reward = 0.
    action = 1
    for i in tqdm.tqdm(range(training_size)):
        optimizer.zero_grad()
        data = torch.from_numpy(train_x[i].reshape(batch_size, data_length, 1).astype(np.float32)).to(device)
        label = torch.from_numpy(train_t[i].astype(np.float32)).to(device)

        output = model(data)

        loss = criterion(output, label)
        loss.backward()
        optimizer.step()

        running_loss += loss.data
        if output.data > 0: #V(s)>0で買い
            action = 0
            buy_chart = m_data[i]
        else: #V(s)<0で売り
            if action == 0: 
                reward += m_data[i] - buy_chart #買った時と売った時の差がreward             
            action = 1

    print('%d loss: %3f, training_accuracy: %.5f' % (epoch+1, running_loss, reward))
```
